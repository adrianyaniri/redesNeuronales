```python
import numpy as np

def generar_conjunto_xor(n_ejemplos):
    x1 = np.random.randint(0, 2, n_ejemplos)
    x2 = np.random.randint(0, 2, n_ejemplos)
    y = np.logical_or(x1, x2) & np.logical_not(np.logical_and(x1, x2))
    return x1, x2, y

x1, x2, y = generar_conjunto_xor(1000)


```

```python
from matplotlib import pyplot as plt

plt.scatter(x1[y == 0], x2[y == 0], label="Clase 0")
plt.scatter(x1[y == 1], x2[y == 1], label="Clase 1")

# Agregar etiquetas
plt.xlabel("x1")
plt.ylabel("x2")

# Agregar leyenda
plt.legend()

# Mostrar la gráfica
plt.show()
```

```python
import numpy as np

def generar_conjunto_xor(n_ejemplos):
    x1 = np.random.randint(0, 2, n_ejemplos)
    x2 = np.random.randint(0, 2, n_ejemplos)
    y = np.logical_or(x1, x2) & np.logical_not(np.logical_and(x1, x2))
    return x1, x2, y

x1, x2, y = generar_conjunto_xor(100)

# Convertir puntos a 1 o -1
x1 = x1 - 0.5
x1 = x1 * 2
x2 = x2 - 0.5
x2 = x2 * 2

# Graficar
import matplotlib.pyplot as plt

plt.scatter(x1[y == 0], x2[y == 0], label="Clase 0")
plt.scatter(x1[y == 1], x2[y == 1], label="Clase 1")

# Agregar etiquetas
plt.xlabel("x1")
plt.ylabel("x2")

# Agregar leyenda
plt.legend()

# Mostrar la gráfica
plt.show()
```
